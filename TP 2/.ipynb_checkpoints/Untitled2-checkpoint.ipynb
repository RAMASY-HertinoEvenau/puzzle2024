```python
from sklearn.cluster import KMeans
import numpy as np

# Préparation des données
X = df.drop(columns=['Region'])

# Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Choisir le nombre optimal de clusters
optimal_clusters = 3  # à déterminer à partir du graphique
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
clusters = kmeans.fit_predict(X)
df['Cluster'] = clusters
```

```python
interpretation = """
Les clusters obtenus représentent des regroupements de régions avec des caractéristiques similaires en termes de taux d'achèvement et de quintiles
de bien-être économique. Les regroupements significatifs peuvent être interprétés en fonction des caractéristiques des clusters.
"""
print(interpretation)

# Visualisation des clusters
plt.scatter(df['Total_Masculin'], df['Total_Feminin'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Total Masculin')
plt.ylabel('Total Feminin')
plt.title('Clusters des régions')
plt.show()
```
